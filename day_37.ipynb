```python
import numpy as np
import pandas as pd

# Reproducible randomness
np.random.seed(45)

# Generate experience data
years = np.random.uniform(0.5, 10, 100).round(2)

# Generate salary data
salaries = (
    30000
    + years * 6000
    + np.random.normal(0, 4000, size=100)
).round(2)

# Create dataframe
df = pd.DataFrame({
    "YearsExperience": years,
    "Salary": salaries
})

# Save CSV
df.to_csv("experience_salary.csv", index=False)

print("Data Saved in file ✅")
```
